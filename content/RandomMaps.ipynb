```python
%matplotlib inline
import copy
import pandas as pd
import pysal as ps
import numpy as np
from pysal.contrib.viz import mapping as maps
import matplotlib.pyplot as plt
from pyGDsandbox.dataIO import dbf2df
```

```python
link = '/Users/dani/Desktop/Houston_mini_course/houston_tract00_rev.dbf'
```

```python
db = dbf2df(link)
db.info()
```

```python
np.random.seed(1234)
shp = ps.open(link.replace('.dbf', '.shp'))
y = np.log(db['AREA'].values)
ry = copy.copy(y)
np.random.shuffle(ry)
```

```python
f = plt.figure(figsize=(12,6))

ax1 = f.add_subplot(121)
ax1.hist(y, alpha=0.5)
ax1.set_title('Var. A')

ax2 = f.add_subplot(122)
h = ax2.hist(ry, alpha=0.5)
ax2.set_title('Var. B')

plt.savefig('/Users/dani/Desktop/hists.png')
plt.show()
```

```python
f = plt.figure(figsize=(12,6))

ax1 = f.add_subplot(121)
polys = maps.map_poly_shp(shp)
polys = maps.base_choropleth_classif(polys, y, cmap='Blues')
polys.set_alpha(0.85)
polys.set_edgecolor('0.5')
polys.set_linewidth(0)
ax1 = maps.setup_ax([polys], ax=ax1)
ax1.set_title('Var. A')

ax2 = f.add_subplot(122)
polys = maps.map_poly_shp(shp)
polys = maps.base_choropleth_classif(polys, ry, cmap='Blues')
polys.set_alpha(0.85)
polys.set_edgecolor('0.5')
polys.set_linewidth(0)
ax2 = maps.setup_ax([polys], ax=ax2)
ax2.set_title('Var. B')

plt.savefig('/Users/dani/Desktop/maps.png')
plt.show()
```
